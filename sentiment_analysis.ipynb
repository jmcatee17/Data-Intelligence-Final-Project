```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Read in Transcript Data
df_transcripts = pd.read_csv('df_transcripts.csv')

df_transcripts
```

```python

# Initialize VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

def analyze_sentiment(text):
    """
    Analyze sentiment of text and return positive, negative, or neutral.
    
    Args:
        text: String to analyze
        
    Returns:
        String: 'positive', 'negative', or 'neutral'
    """
    if pd.isna(text) or text == '':
        return 'neutral'
    
    # Get sentiment scores
    scores = analyzer.polarity_scores(str(text))
    compound_score = scores['compound']
    
    # Classify based on compound score
    # Compound score ranges from -1 (most negative) to +1 (most positive)
    if compound_score >= 0.05:
        return 'positive'
    elif compound_score <= -0.05:
        return 'negative'
    else:
        return 'neutral'

# Apply sentiment analysis to full_transcript column
print("Analyzing sentiment for transcripts...")
df_transcripts['sentiment'] = df_transcripts['full_transcript'].apply(analyze_sentiment)

# Display results
print(f"\nTotal transcripts analyzed: {len(df_transcripts)}")
print("\nSentiment Distribution:")
print(df_transcripts['sentiment'].value_counts())
print("\nSentiment Percentages:")
print(df_transcripts['sentiment'].value_counts(normalize=True).mul(100).round(2))

# Show sample results
print("\nSample results:")
print(df_transcripts[['transcriptid', 'ticker', 'sentiment']].head(10))
```
